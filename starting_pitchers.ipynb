```python
import pandas
from sklearn import preprocessing
```

```python
pitchers = pandas.read_csv("data/FantasyPros_2019_Projections_SP.csv")
```

```python
stats = ["K", "ERA", "WHIP", "QS", "SV+HLD"]
normalised_stats = ["{}norm".format(col) for col in stats]
```

```python
pitchers["ERA"] = 1 / pitchers["ERA"]
pitchers["WHIP"] = 1 / pitchers["WHIP"]
pitchers["SV+HLD"] = 0 
```

```python
# Normalise
x = pitchers[stats].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(pitchers[stats])
normalised = pandas.DataFrame(x_scaled, columns=pitchers[stats].columns)
pitchers = pitchers.join(normalised, rsuffix="norm")
```

```python
pitchers["score"] = pitchers[normalised_stats].mean(axis=1)
```

```python
pitchers[["Player", "Own", "Team", "Positions", "score"] + stats + normalised_stats].sort_values("score", ascending=False)
```
