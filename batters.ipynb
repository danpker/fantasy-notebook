```python
import pandas
from sklearn import preprocessing
```

```python
batters = pandas.read_csv("data/batters_zips.csv")
players = pandas.read_csv("data/players.csv")
```

```python
batters['playerid'] = batters['playerid'].astype(str)
players['fg_id'] = players['fg_id'].astype(str)
batters = batters.merge(players, left_on="playerid", right_on="fg_id")
```

```python
stats = ["R", "HR", "RBI", "SB", "OBP"]
normalised_stats = ["{}norm".format(col) for col in stats]
```

```python
# Normalise
x = batters[stats].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(batters[stats])
normalised = pandas.DataFrame(x_scaled, columns=batters[stats].columns)
batters = batters.join(normalised, rsuffix="norm")
```

```python
batters["score"] = batters[normalised_stats].mean(axis=1)
```

```python
batters[["Name", "mlb_pos", "yahoo_pos", "ADP", "score"] + normalised_stats].sort_values("score", ascending=False)
```
